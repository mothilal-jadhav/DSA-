```python
# lets filter an array lets say that i only want odd numbers
k = 0
for i in arr:
    if i%2 != 0:
        arr[k] = i
        k += 1

print(arr[:k]) # here the arr would be [1,3,11,11,10], so the list of odd i have sliced the array else we can also do
```

```python
arr = arr[:k]
print(arr) # now our array is filtered with only odd numbers
```

```python
# i can also fetch a sub array with slicing

arr1 = arr[:2]
arr2 = arr[1:]
print(arr1)
print(arr2)
```

```python
# fetching all subarrays by loop

for i in range(len(arr)):
    for j in range(i+1,len(arr)+1):
        sub_arr = arr[i:j]
        print(sub_arr) # for a len of n the no of possible sub arrays will be n(n+1)/2, for our example it will be 6 sub arrays -> [1],[1,3],[1,3,11],[3],[3,11],[11]
```

```python
# since we have arr1 and arr2 as two different arrays as we sliced earlier lets merge them both

new_arr = arr1+arr2
print(new_arr)
```

```python
# we can also merge two arrays based on their values as we want as ascending or descending 

odd = [1,3,5,7,9,11] # all odd numbers
even = [2,4,6,8,10] # all even numbers

# lets merge them and return an array of natural numbers in increasing order from 1-11

n1 = len(odd)
n2 = len(even)

i = 0
j = 0
natural = []

while i<n1 and j<n2:
    if odd[i]<=even[j]:
        natural.append(odd[i])
        i+=1
    else:
        natural.append(even[j])
        j += 1

natural.extend(odd[i:])
natural.extend(even[j:])

print(f'the odd numbers are {odd}')
print(f'the even numbers are {even}')
print(f'after merging these two the sorted natural numbers are {natural}')
```

```python
# we can also rotate an array and it will still be in non decreasing order

# lets say we want to rotate the array by two positions

d = 2 # can also take it as input or use it as our convinience

for _ in range(d):
    last = odd[n1-1]

    for i in range(n1-1,0,-1):
        odd[i] = odd[i-1]

    odd[0] = last

print(odd) # time complexity will be O(n*d) and if d is large then too much hectic
```
