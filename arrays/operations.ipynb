```python
# declaration of array

arr = []
```

```python
# initialising an array

arr = [1,2,3,4,5,'mothilal',(1,2,3),[1,2,3,4]]

# can have any data types
```

```python
# fixed type arrays

fixed_arr = [0]*5
print(fixed_arr)
```

```python
# dynamic array
arr = []
print(arr)
```

```python
# lets iterate through an array

arr = [1,2,3,4,5,6]
for i in arr:
    print(i,end=" ")
```

```python
# lets iterate from reverse

for i in range(len(arr)-1, -1, -1):
    print(arr[i], end=' ')
```

```python
# if i want to get an element say 3rd element

print(arr[2]) # 3rd element is 2nd index as the counting start from 0
```

```python
# search an element, if exists then return True else False

for i in arr:
    if i==3:
        print(True)
```

```python
# insertion of elements

print(arr) # before insertion
arr.insert(0,8) # insertion at beginning using built method
print(arr) # after insertion
arr[0] = 9 # insertion at beginning using custom method
print(arr)#after insertion
arr.append(10) #insertion at end using built method
print(arr)
arr[10] = 11 # insertion at beginning using custom method
print(arr)
```

```python
# deletion of elements

print(arr)
del arr[0] #deletion of first element
print(arr)
del arr[2] # deletion at required position
print(arr)
arr.remove(5) # built in way of deletinge elements
print(arr)
```

```python
# lets filter an array lets say that i only want odd numbers
k = 0
for i in arr:
    if i%2 != 0:
        arr[k] = i
        k += 1

print(arr[:k]) # here the arr would be [1,3,11,11,10], so the list of odd i have sliced the array else we can also do
```

```python
arr = arr[:k]
print(arr) # now our array is filtered with only odd numbers
```

```python
# i can also fetch a sub array with slicing

arr1 = arr[:2]
arr2 = arr[1:]
print(arr1)
print(arr2)
```

```python
# fetching all subarrays by loop

for i in range(len(arr)):
    for j in range(i+1,len(arr)+1):
        sub_arr = arr[i:j]
        print(sub_arr) # for a len of n the no of possible sub arrays will be n(n+1)/2, for our example it will be 6 sub arrays -> [1],[1,3],[1,3,11],[3],[3,11],[11]
```
