```python
# dynamic array
arr = []
print(arr)
```

```python
# lets iterate through an array

arr = [1,2,3,4,5,6]
for i in arr:
    print(i,end=" ")
```

```python
# lets iterate from reverse

for i in range(len(arr)-1, -1, -1):
    print(arr[i], end=' ')
```

```python
# if i want to get an element say 3rd element

print(arr[2]) # 3rd element is 2nd index as the counting start from 0
```

```python
# search an element, if exists then return True else False

for i in arr:
    if i==3:
        print(True)
```
