```python
# lets create a linked list

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None # classic Node strucuture of Linked List


```

```python
# creting the first node
Head = Node(10)

# noe the next nodes will be its inserted sequentially next to its pointer
Head.next = Node(20)
Head.next.next = Node(30)
Head.next.next.next = Node(40)

temp = Head

while temp is not None:
    print(temp.data,end = " -> ")
    temp = temp.next
    
```

```python
# traversing through linked list 

def traverse(head):
    while head is not None:
        print(head.data, end='')
        if head.next is not None:
            print('->', end='')
        head = head.next

    print()

print(traverse(Head))
```

```python
# lets insert a node at beggining

def insertBegi(head,x):
    newNode = Node(x)
    newNode.next = head

    return newNode

Head = insertBegi(Head,5)
print(traverse(Head))
```

```python
# insert the element at the end

def insertEnd(head,x):
    newNode = Node(x)
    
    if head is None:
        return newNode
    
    last = head

    while last.next is not None:
        last = last.next

    last.next = newNode

    return head

Head = insertEnd(Head,50)
print(traverse(Head))
```

```python
# insert at desired position

def insert(head,val,pos):
    if pos<1:
        return head
    
    if pos==1:
        head = insertBegi(head,val)
        return head
    
    curr = head

    for i in range(1,pos-1):
        if curr is None:
            return head
        
        curr = curr.next

    if curr is None:
        return head
    
    newNode = Node(val)

    newNode.next = curr.next
    curr.next = newNode

    return head

Head = insert(Head,15,3)
print(traverse(Head))
```

```python
# lets delete elements starting from the first position

def deleteBegi(head):
    if head is None:
        return None
    
    temp = head

    head = head.next

    temp = None

    return head

Head = deleteBegi(Head)
print(traverse(Head))
```

```python
# delete at last position

def deleteLast(head):
    if head is None:
        return None
    
    if head.next is None:
        return None

    secondLast = head
    
    while secondLast.next.next is not None:
        secondLast = secondLast.next


    secondLast.next = None

    return head

Head = deleteLast(Head)
print(traverse(Head))
    


    
```

```python
# deletion at desired position

def delete(head,pos):
    temp = head

    if pos == 1:
        head = temp.next 
        return head
    
    prev = None

    for i in range(1,pos):
        prev = temp
        temp = temp.next

    prev.next = temp.next 
    return head

Head = delete(Head,4)
print(traverse(Head))
```

```python
# lets search the element we want 

def searchEle(head,x):
    if head is None:
        return False
    
    while head is not None:
        if head.data == x:
            return True
        
        head = head.next
    return False

print(searchEle(Head,15))
print(searchEle(Head,13))
```

```python
# reverse the linked list

def reverse(head):
    if head is None:
        return head
    
    curr = head
    prev = None

    while curr is not None:
        nextNode = curr.next
        curr.next = prev
        prev = curr
        curr = nextNode
    return prev

Head = reverse(Head)
print(traverse(Head))
```
