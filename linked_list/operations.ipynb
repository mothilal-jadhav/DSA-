```python
# lets create a linked list

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None # classic Node strucuture of Linked List


```

```python
# creting the first node
Head = Node(10)

# noe the next nodes will be its inserted sequentially next to its pointer
Head.next = Node(20)
Head.next.next = Node(30)
Head.next.next.next = Node(40)

temp = Head

while temp is not None:
    print(temp.data,end = " -> ")
    temp = temp.next
    
```

```python
# traversing through linked list 

def traverse(head):
    while head is not None:
        print(head.data, end='')
        if head.next is not None:
            print('->', end='')
        head = head.next

    print()

print(traverse(Head))
```

```python
# lets insert a node at beggining

def insertBegi(head,x):
    newNode = Node(x)
    newNode.next = head

    return newNode

Head = insertBegi(Head,5)
print(traverse(Head))
```

```python
# insert the element at the end

def insertEnd(head,x):
    newNode = Node(x)
    
    if head is None:
        return newNode
    
    last = head

    while last.next is not None:
        last = last.next

    last.next = newNode

    return head

Head = insertEnd(Head,50)
print(traverse(Head))
```
