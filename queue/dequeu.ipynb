```python
from collections import deque

q = deque()
print(q)
```

```python
q.append(10) # append adds the value into the dequeue in O(n) order
print(q)
```

```python
q. append(12) # append adds the value on to the right side of the dequeue
print(q)
```

```python
q.appendleft(8)  # appendleft is used to add the value at the left side of the dequeue
print(q)
```

```python
q.extend([14,16,18]) # extend adds the value onto the right side of the deque
print(q)
```

```python
q.extendleft([6,4,2]) # extend left adds the elements on to the left in reverse order
print(q)
```

```python
q.remove(12) # remove function takes a value and removes that element from the deque if value not found then it raises Valuerror
print(q)
q.remove(19)
print(q)
```

```python
q.pop() # removes the rear element from the deque
print(q)
```

```python
q.popleft() # removes the leftmost element that is front element
print(q)
```

```python
# dequeue is a double ended queue which allows us to add and delete elements from both the sides,
# it can be used in mostly sliding window problems
# it is also used in task scheduling and real time data processing 
```
